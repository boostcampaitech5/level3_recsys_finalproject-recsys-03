```python
from collections import Counter
```

```python
from src.utils import str2list
```

```python
matching_df = pd.DataFrame(
    {'playlist_id' :Counter(id_list).keys(),
    'matched' : Counter(id_list).values() }
    )
matching_df
```

```python
pl_result = playlist.loc[playlist['playlist_id'].isin(id_list)][['playlist_id','playlist_songs','num_of_songs','playlist_view', 'playlist_likecount','tag_sit_cnt', 'tag_mood_cnt', 'tag_weather_cnt' ]]
pl_result = pl_result.reset_index(drop = True)
```

```python
pl_result['tag_sit_cnt'] = pl_result['tag_sit_cnt'].apply(lambda x: min(x, 1))
pl_result['tag_mood_cnt'] = pl_result['tag_mood_cnt'].apply(lambda x: min(x, 1))
pl_result['tag_weather_cnt'] = pl_result['tag_weather_cnt'].apply(lambda x: min(x, 1))
```

```python
pl_result['tag_cnt'] = pl_result.tag_sit_cnt + pl_result.tag_mood_cnt + pl_result.tag_weather_cnt
```

```python
str2list(pl_result, ['playlist_songs'])
```

```python
pl_result
```

```python
pl = pd.merge(matching_df, pl_result)
pl['match_rate'] = pl.matched/pl.tag_cnt
```

```python
pl = pl.drop(["matched", "tag_sit_cnt", "tag_mood_cnt", "tag_weather_cnt", "tag_cnt"], axis=1)
pl
```

```python
songs = []
for song_list in pl.playlist_songs:
    songs += song_list
songs = list(set(songs))
```

```python
pl_id = []
song_id = []
pl_view = []
pl_like = []
pl_match = []

for id in pl.playlist_id:
    i, song_list, _, v, l, mr = pl[pl.playlist_id == id].values.flatten()
    for song in song_list:
        pl_id.append(i)
        song_id.append(song)
        pl_view.append(v)
        pl_like.append(l)
        pl_match.append(mr)
```

```python
pl_song = pd.DataFrame(
    {"playlist_id" : pl_id,
    "song_id" : song_id,
    "pl_view": pl_view,
    "pl_like" : pl_like,
    "pl_match" : pl_match}
)
```

```python
def min_max_scale(df, col:str):
    min_val = min(df[col])
    max_val = max(df[col])
    df[col] = df[col].apply(lambda x: round((x-min_val)/(max_val-min_val), 2))
```

```python
min_max_scale(merged, "pl_favor")
merged
```

```python
min_max_scale(merged, "song_favor")
```

```python
merged['favor_score'] = merged.song_favor + merged.pl_favor
merged
```
